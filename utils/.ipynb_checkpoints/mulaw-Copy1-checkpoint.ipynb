```python
import numpy as np
from mulaw import mulaw

from numpy import *
import math
import matplotlib.pyplot as plt
```

```python
numdvals=10
mlaw= mulaw(numdvals)

#Testing ground

# floats->encoded->decimated->decoded 
orig = np.linspace(-1, 1, num=numdvals, endpoint=True)
print("orig = " + str(orig))
mu=mlaw._float2mu(orig)
print("mu = " + str(mu))
z=mlaw._decimate(mu)
print("z = " + str(z))
f=mlaw._mu2float(z)
print("f = " + str(f))

# check one-hot coding 
print("\n---  check one-hot coding ---")
oh=np.zeros(numdvals)
oh[int(numdvals/2)]=1
print(oh)
om = mlaw.decode([oh])
print("om = " + str(om))
 
idx=mlaw._mu2index(0)
print("idx = " + str(idx) + " (should be numdvals/2 -1) for even number of vals")
```

```python
t = linspace(0,2*math.pi,400)
a = sin(t)
# encode 
oh = mlaw.encode(a)
print(str(shape(oh)))
c=mlaw.decode(oh)

plt.plot(t, a, t, c)
plt.show()
```

```python
# Check and visualize

t = linspace(0,2*math.pi,400)
a = sin(t)
# encode then decode
b = mlaw._mu2float(mlaw._decimate(mlaw._float2mu(a)))
# test one hot conversions  (the descretized b and c should look exactly the same)
c = mlaw._mu2float(mlaw._onehot2mu(mlaw._mu2onehot(mlaw._float2mu(a))))
# decimate mulaw
d = mlaw._float2mu(a)

fig = plt.figure()
ax = plt.subplot(111)
ax.plot(t, a, label='sin')
ax.plot(t, b, label='sin->mu->decim->decode')
ax.plot(t, c, label='sin->mu->1hot->mu->decode')
ax.plot(t, d, label='sin->mu')
plt.title('mu-law encoding')
ax.legend()
plt.show()
```
